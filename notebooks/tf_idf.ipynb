```python
def count_df(text, use_nltk=True):
    freq = {}
    if use_nltk:
        full_list = wordpunct_tokenize(text)
    else:
        full_list = []
        for x in text.split('\n'):
            full_list.extend(x.split(' '))

    for word in full_list:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
```
