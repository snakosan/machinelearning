```python
def process_bios_e820(line):
    expr = '\[mem[\s]*(\S+)-(\S+)\]\s(\S+)'
    return  re.findall(expr, line)[0]

for line in log_text.split('\n'):
    if 'BIOS-e820' in line:
        #print(line)
        print(process_bios_e820(line))
    if 'ACPI' in line:
        print(line)

```

```python
def extract_section(log_text, section_tag):
    res = []
    for line in log_text.split('\n'):
        if section_tag in line.upper():
            res.append(line)
    return res
        
```
